#Customer Churn Prediction
## Data Understanding
Objective
The objective of this notebook is to understand the IBM Telco Customer Churn dataset before performing exploratory data analysis and machine learning.

We inspect the dataset structure, identify missing values, understand data types, analyze the target variable, and document initial observations

```python
#Import Libraries

# Data Manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option("display.max_columns", None)

# Plot settings
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)

```

```python
#Load Dataset
df = pd.read_csv("Customer_Churn_Dataset.csv")
```

```python
#Display First 5 Rows
df.head()
```

```python
#Shape of Dataset
print("Rows    :", df.shape[0])
print("Columns :",df.shape[1])
```

```python
#Column Names
df.columns.tolist()
```

```python
#Dataset Information
df.info()
```

```python
#Statistical Summary
df.describe()
```

```python
#Categorical Summary
df.describe(include="object")
```

```python
#Missing Values
df.isnull().sum()
```

```python
#Hidden Missing Values
(df["TotalCharges"] == " ").sum()
```

```python
#Duplicate Records
print("Duplicate Rows :", df.duplicated().sum())
```

```python
#Target Variable
df["Churn"].value_counts()

```

```python
(df["Churn"].value_counts(normalize=True)*100).round(2)
```

```python
#Visualize the Target
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="Churn",
)

plt.title("Customer Churn Distribution")

plt.xlabel("Churn")

plt.ylabel("Count")

plt.show()
```

# Key Takeaways

• Dataset contains 7043 customer records.

• There are 21 features.

• No duplicate records exist.

• TotalCharges contains blank strings.

• Churn classes are moderately imbalanced.

• Dataset is suitable for binary classification.
