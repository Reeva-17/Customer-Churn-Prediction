```python
import pandas as pd

lr = pd.read_csv("reports/logistic_regression_results.csv")
dt = pd.read_csv("reports/decision_tree_results.csv")
svm = pd.read_csv("reports/svm_results.csv")
rf = pd.read_csv("reports/random_forest_results.csv")
xgb = pd.read_csv("reports/xgboost_results.csv")
```

```python
# Load Processed Dataset
X_train = pd.read_csv("data/processed/X_train.csv")
X_test = pd.read_csv("data/processed/X_test.csv")

y_train = pd.read_csv("data/processed/y_train.csv").squeeze()
y_test = pd.read_csv("data/processed/y_test.csv").squeeze()
```

```python
comparison = pd.concat(
    [lr, dt, svm,rf, xgb],
    ignore_index=True
)

comparison
```

```python
comparison.to_csv(
    "reports/model_comparison.csv",
    index=False
)
```

```python
best_model = comparison.loc[
    comparison["F1 Score"].idxmax()
]

print(best_model)
```

```python
best_model_name = best_model["Model"]

print("Best Model:", best_model_name)
```
